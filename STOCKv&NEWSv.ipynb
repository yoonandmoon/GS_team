```python
import pandas as pd
import matplotlib.pyplot as plt

#  
df = pd.read_csv('/content/R_gs_stock(ver.210801_231027).csv')
df2 = pd.read_csv('/content/lasso.csv')

# 'date'  datetime  
df['date'] = pd.to_datetime(df['date'])

# X :  ,  Y :  
plt.figure(figsize=(5, 6))
plt.plot(df['date'], df[''], color='k', label='stock v',linestyle='-')
plt.xlabel('date')
plt.ylabel('stock v')
plt.xticks(rotation=45)  # X   45  

#  Y :  ()
ax2 = plt.twinx()
ax2.plot(df['date'], df2['count'], color='r', label='news v', linestyle='--')
ax2.set_ylabel('news v')

#  
lines1, labels1 = plt.gca().get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
lines = lines1 + lines2
labels = labels1 + labels2
plt.legend(lines, labels, loc='upper left')

plt.title('STOCK v & NEWS v')

plt.show()
```
