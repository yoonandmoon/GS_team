```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터프레임 불러오기
df = pd.read_csv('/content/R_gs_stock(ver.210801_231027).csv')
df2 = pd.read_csv('/content/lasso.csv')

# 'date' 열을 datetime 형식으로 변환
df['date'] = pd.to_datetime(df['date'])

# X 축: 날짜 데이터, 왼쪽 Y 축: 주식 거래량
plt.figure(figsize=(5, 6))
plt.plot(df['date'], df['거래량'], color='k', label='stock v',linestyle='-')
plt.xlabel('date')
plt.ylabel('stock v')
plt.xticks(rotation=45)  # X 축 라벨을 45도 회전하여 표시

# 오른쪽 Y 축: 뉴스량 (가정)
ax2 = plt.twinx()
ax2.plot(df['date'], df2['count'], color='r', label='news v', linestyle='--')
ax2.set_ylabel('news v')

# 범례 표시
lines1, labels1 = plt.gca().get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
lines = lines1 + lines2
labels = labels1 + labels2
plt.legend(lines, labels, loc='upper left')

plt.title('STOCK v & NEWS v')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터프레임 불러오기
df = pd.read_csv('/content/R_gs_stock(ver.210801_231027).csv')
df2 = pd.read_csv('/content/lasso.csv')

# 산점도 그래프
plt.figure(figsize=(8, 6))
plt.scatter(df['거래량'], df2['count'], c='k', alpha=0.5)  # c는 색상, alpha는 투명도
plt.xlabel('s v')
plt.ylabel('n v')
plt.title('s & n')
plt.grid(True)

plt.show()
```
