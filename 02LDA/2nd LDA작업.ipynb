```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/gs건설 프로젝트/news_combined2.csv')
df
```

```python
from gensim import corpora, models
from pprint import pprint

df['tokens'] = df['tokens'].fillna('')
df['tokens'] = df['tokens'].astype(str)
documents = df['tokens'].values.tolist()
documents
```

```python
tokenized_documents = [str(doc).split() for doc in documents]
dictionary = corpora.Dictionary(tokenized_documents)
corpus = [dictionary.doc2bow(doc) for doc in tokenized_documents]
num_topics = 30  # 주제의 수를 지정
lda_model = models.LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=15)
```

```python
topics = lda_model.print_topics(num_words=5)
for topic in topics:
    print(topic)
```

## 뉴스 데이터 분류

```python
df['dominant_topic'] = None
df['dominant_topic_probability'] = None
df
```

```python
for i, row in df.iterrows():
    preprocessed_doc = [w for w in row['tokens'].split(' ') if len(w) > 1]
    bow = dictionary.doc2bow(preprocessed_doc)
    topic_info = lda_model[bow]
    # 지배적인 주제를 선택할 때 확률이 가장 높은 주제를 선택
    dominant_topic = max(topic_info, key=lambda x: x[1])
    df.at[i, 'dominant_topic'] = dominant_topic[0]
    df.at[i, 'dominant_topic_probability'] = dominant_topic[1]
```
