```python
from gensim import corpora, models
from pprint import pprint

df['Tokens_2'] = df['Tokens_2'].fillna('')
df['Tokens_2'] = df['Tokens_2'].astype(str)
df['Title'] = df['Title'].fillna('')
df['Title'] = df['Title'].astype(str)
documents = df['Tokens_2'].values.tolist()
documents
#3
```

```python
titles = df['Title'].tolist()
titles
#4
```

```python
tokenized_documents = [str(doc).split() for doc in documents]
#5
```

```python
title_tokens = [str(title).split() for title in titles]
#6
```

```python
dictionary = corpora.Dictionary(tokenized_documents)
#7
```

```python
import gensim
from gensim.models import TfidfModel
```

```python
tfidf_model = TfidfModel([dictionary.doc2bow(text) for text in title_tokens])
#8
```

```python
combined_tokens = [title_tokens[i] + tokenized_documents[i] for i in range(len(title_tokens))]
tfidf_tokens = [tfidf_model[dictionary.doc2bow(text)] for text in combined_tokens]
```

```python
num_topics = 50  # 주제의 수를 지정
lda_model = models.LdaModel(corpus=tfidf_tokens, num_topics=num_topics, id2word=dictionary, passes=15)
lda_model
```

```python
# 주제 및 주제 단어 출력
topics = lda_model.print_topics(num_words=6)
for topic in topics:
    print(topic)
#9
```

```python
df['dominant_topic'] = None
df['dominant_topic_probability'] = None
df2 = df[['date','dominant_topic','dominant_topic_probability']]
df2['token'] = df['Title']+ df['Tokens_2']
df2
#10
```

```python
df2
```

```python
# !pip install tqdm
```

```python
for i, row in df2.iterrows():
    preprocessed_doc = [w for w in row['token'].split(' ') if len(w) > 1]
    bow = dictionary.doc2bow(preprocessed_doc)
    topic_info = lda_model[bow]
    # 지배적인 주제를 선택할 때 확률이 가장 높은 주제를 선택
    dominant_topic = max(topic_info, key=lambda x: x[1])
    df2.at[i, 'dominant_topic'] = dominant_topic[0]
    df2.at[i, 'dominant_topic_probability'] = dominant_topic[1]
#11
```

```python
# pyLDAvis 시각화
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim.prepare(lda_model, tfidf_tokens, dictionary, sort_topics=False)  # sort_topics 매개변수를 비활성화
pyLDAvis.display(vis)
#12
```

```python
top_words_per_topic = []
for t in range(lda_model.num_topics):
    top_words_per_topic.extend([(t,) + x for x in lda_model.show_topic(t, topn=6)])
# 추출된 단어들을 데이터프레임으로 변환합니다.
df_top_words = pd.DataFrame(top_words_per_topic, columns=['Topic', 'Word', 'Probability'])
# 결과를 확인합니다.
df_top_words

#13
```

```python
df_top_words.to_csv('lda_topic_stop(v.jw.1).csv', index=False)
#14
```
