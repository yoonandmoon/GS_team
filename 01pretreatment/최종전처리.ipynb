```python
stop_word = ['분양', '가구', '국내', '브랜드', '관계자', '공급', '제공', '건설사', '입주', '면적', '기록', '사진',
            '지상', '순위', '경우', '물량', '부문', '세대', '서비스','건설', '청약', '전용', '주택','공원', '시작','수도','마련','생활','센터',
            '대상','생산','평균','지구','회사','확대','컨소시엄']

def preprocess(text):
  text = text.split()
  text = [i for i in text if i not in stop_word]
  return text
```

```python
def make_tokens(df):
    df['Tokens_2'] = ' '
    for i, row in df.iterrows():
        if i % 100 == 0:
            print(i, '/', len(df))
        tokens = df['tokens'][i]

        if isinstance(tokens, str):
            alpha = preprocess(tokens)
            df['Tokens_2'][i] = ' '.join(alpha)

    return df

df2 = make_tokens(df)
df2
```

```python
df3 = df2[['date','Title','Content','Tokens_2']]
df3
```

```python
from collections import Counter

# 'tokens' 열에서 NaN 또는 float 값을 건너뛰기
words_list = df3['Tokens_2'].apply(lambda x: str(x).split())

# 모든 단어를 얻기 위해 중첩된 리스트를 평평하게 만듭니다.
all_words = [word for words in words_list for word in words]

word_counts = Counter(all_words)

most_common_words = word_counts.most_common(100)
print("가장 많이 나온 상위 100개 단어:")
for word, count in most_common_words:
    print(f"{word}: {count}")
```
